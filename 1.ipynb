## Instalar miniconda

```python
Invoke-WebRequest -Uri "https://repo.anaconda.com/miniconda/Miniconda3-latest-Windows-x86_64.exe" -outfile ".\miniconda.exe"
```

```python
Start-Process -FilePath ".\miniconda.exe" -ArgumentList "/S" -Wait
```
